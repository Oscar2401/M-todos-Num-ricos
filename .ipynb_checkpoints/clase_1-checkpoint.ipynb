```python
import numpy as np
from pylab import*
from matplotlib import*
```

# Graficar

```python
x=np.linspace(0,20,500) #array de 0 a 20 con 50 datos
y=np.sin(x)
print(x)
print(y)
```

```python
#python color chart
x=np.linspace(0,20,50) #array de 0 a 20 con 50 datos
y=np.sin(x)
print(x)
print(y)
plot(x,y,'--*',color = 'salmon')
xlabel('x')
ylabel('sin(x)')
pylab.savefig('sin(x).pdf')
show()
```

#### Ejercicio

```python
x=np.linspace(-5,5,500)
y=np.exp(-x**2)/(np.pi)**(1/2) 
plot(x,y,'-',color = 'magenta')
xlabel('x')
ylabel('$\\frac{e^{-x^2}}{\sqrt{\pi}}$')
show()
```

```python
#-pi a pi
x = np.linspace(-np.pi,np.pi,500)
y = np.cos(x) + np.sin(x)
plot(x,y,'-',color = 'navy')
xlabel('x')
ylabel('cos(x) + sin(x)')
show()
```

# Definir Funciones

```python
def g(x):
    return x**2 + x*exp(x)

x = np.linspace(-np.pi,np.pi,500)
plot(x,g(x))
plot(x,sin(x))
#ylim(0.,5)
show()
```
