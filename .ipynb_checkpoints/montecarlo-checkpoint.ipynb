```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pylab as pl 
```

```python
x_0 = 2.0 #Mean (“centre”) of the distribution.
dev = 0.4 #Standard deviation (spread or “width”) of the distribution.
D = 1
```

```python
s = np.random.normal(x_0,dev, 10000)
```

```python
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(dev * np.sqrt(2 * np.pi)) * np.exp( - (bins - x_0)**2 / (2 * dev**2) ),linewidth=3, color='r')
plt.show()
```

```python
def dif(t,x,D):
    N = (1/np.sqrt(4*np.pi*D*t))
    return N*np.exp(-np.divide(np.power(x,2),4*D*t))
```

```python
t = np.linspace(1,32,5)
t_0 = 0
x_0 = 2
D =1
dev = (2*D*t)**(1/2) 
s = []
for i in range(len(dev)):
    s.append(np.random.normal(x_0,dev[i], 1000))
```

```python
x = [x_0]
for k in range(len(dev)):
    for i in range(1,len(s[0])):
        x.append(x[0] + s[k][i]*math.sqrt(2*D*(t[k]-t_0)))
```

```python
f = []
for i in range(len(t)):
    f.append(dif(t[i],x,D))
```

```python
for i in range(len(f)):
    plt.plot(x,f[i],'.',label='$ t = $')
    plt.xlim(-20,20)
plt.legend(loc='center right', shadow=True, frameon = False , fontsize='x-large')
```
