# Método de Punto fijo 

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# función f(x)
x = np.linspace(-5,5,500)
y = x**3 + 4*x**2 -10
plt.plot(x,y)
plt.xlabel("x",fontsize=18)
plt.ylabel("f(x)",fontsize=18)
plt.show()

#función g(x) = x
x = np.linspace(-5,5,500)
y = np.divide(np.power(10-np.power(x,3),0.5),2)
plt.plot(x,y)
plt.plot(x,x)
plt.xlabel("x",fontsize=18)
plt.ylabel("g(x),x",fontsize=18)
plt.show()
```

```python
def g(x):
    return np.divide(np.power(10-np.power(x,3),0.5),2)

def p_fig(p_0):
    p_0 = 2
    while abs(g(p_0) - p_0) > 0.0000005:
        p_0 =  g(p_0)
    return p_0 

x = np.linspace(-5,5,500)
y = g(x)
plt.plot(x,y)
plt.plot(x,x)
p = p_fig(2)
print('puntp p=',p)
plt.plot(p,g(p),'ro',label='$raiz={:.3f}$'.format(p))
plt.xlabel("x",fontsize=18)
plt.ylabel("g(x),x",fontsize=18)
plt.legend(loc='upper right', shadow=True, fontsize='x-large')
plt.show()

x = np.linspace(-5,5,500)
y = x**3 + 4*x**2 -10
plt.plot(x,y)
plt.plot(p,p**3 + 4*p**2 -10,'ro',label='$raiz={:.3f}$'.format(p))
plt.xlabel("x",fontsize=18)
plt.ylabel("f(x)",fontsize=18)
plt.legend(loc='upper right', shadow=True, fontsize='x-large')
plt.show()
```

```python
def g(x):
    return np.power(np.cos(x),2)

def p_fig(p_0):
    while abs(g(p_0) - p_0) > 0.0000005:
        p_0 =  g(p_0)
    return p_0 

x = np.linspace(0,1,500)
y = g(x)
plt.plot(x,y)
plt.plot(x,x)
p = p_fig(0.8)
print(p)
plt.plot(p,g(p),'ro',label='$raiz={:.3f}$'.format(p))
plt.xlabel("x",fontsize=18)
plt.ylabel("g(x),x",fontsize=18)
plt.legend(loc='upper right', shadow=True, fontsize='x-large')
plt.show()

x = np.linspace(0,1,500)
y = np.sqrt(x) - np.cos(x)
plt.plot(x,y)
plt.plot(p,np.sqrt(p) - np.cos(p),'ro',label='$raiz={:.3f}$'.format(p))
plt.xlabel("x",fontsize=18)
plt.ylabel("f(x)",fontsize=18)
plt.legend(loc='upper right', shadow=True, fontsize='x-large')
plt.show()
```
