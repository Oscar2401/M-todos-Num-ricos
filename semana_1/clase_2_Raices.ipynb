# Método de Bisección

Encontrar las raices de la función:

\begin{equation}
    f(x) = \sqrt{x} - \cos{x}
\end{equation}

```python
#graficamos para encontrar los puntos cercas de la raiz
x = np.linspace(0,1,500)
y = np.sqrt(x)-np.cos(x)
plt.plot(x,y)
plt.xlabel("x",fontsize=18)
plt.ylabel("f(x)",fontsize=18)
plt.show()
```

```python
a = 0.6
b = 0.8 

if np.sqrt(a)-np.cos(a) < 0 and np.sqrt(b)-np.cos(b) > 0:
    p = (a+b)/2
    for i in range(0,100):
        y_p = np.sqrt(p)-np.cos(p)
        if y_p < 0:
            a = p
        if y_p > 0:
            b = p
        p = (a+b)/2
    
    print(y_p)
    x = np.linspace(0,1,500)
    x_l = np.linspace(0,p,500)
    y_l = np.zeros(len(x_l))
    y = np.sqrt(x)-np.cos(x)
    plt.plot(x,y)
    plt.xlabel("x",fontsize=18)
    plt.ylabel("f(x)",fontsize=18)
    plt.plot(p,y_p,'ro',label='$raiz={:.6f}$'.format(p))
    plt.plot(x_l,y_l,"r--")
    plt.legend(loc='upper right', shadow=True, fontsize='x-large')
    plt.show()

if np.sqrt(a)-np.cos(a) > 0:
    print("El valor de a es erroneo")
if np.sqrt(b)-np.cos(b) < 0:
    print("El valor de b es erroneo")
if np.sqrt(a)-np.cos(a) > 0 and np.sqrt(b)-np.cos(b) < 0:
    print("El valor de a y b son erroneos")
```

Encontrar las raices de la función:

\begin{equation}
    f(x) = x^2 - e^{x}
\end{equation}

```python
#graficamos para encontrar los puntos cercas de la raiz
x = np.linspace(-4,2.5,500)
y = np.power(x,2) - np.exp(x) 
plt.plot(x,y)
plt.xlabel("x",fontsize=18)
plt.ylabel("f(x)",fontsize=18)
plt.show()
```

```python
a = 2
b = -2

if np.power(a,2) - np.exp(a)  < 0 and np.power(b,2) - np.exp(b)  > 0:
    p = (a+b)/2
    for i in range(0,100):
        y_p = np.power(p,2) - np.exp(p) 
        if y_p < 0:
            a = p
        if y_p > 0:
            b = p
        p = (a+b)/2

    print(y_p)    
    x = np.linspace(-4,2.5,500)
    y = np.power(x,2) - np.exp(x) 
    x_l = np.linspace(-4,p,500)
    y_l = np.zeros(len(x_l))
    plt.plot(x,y)
    plt.xlabel("x",fontsize=18)
    plt.ylabel("f(x)",fontsize=18)
    plt.plot(p,y_p,'ro',label='$raiz={:.6f}$'.format(p))
    plt.plot(x_l,y_l,"r--")
    plt.legend(loc='upper right', shadow=True, fontsize='x-large')
    plt.show()
    
if np.power(a,2) - np.exp(a) > 0:
    print("El valor de a es erroneo")
if np.power(b,2) - np.exp(b) < 0:
    print("El valor de b es erroneo")
if np.power(a,2) - np.exp(a) > 0 and np.power(b,2) - np.exp(b) < 0:
    print("El valor de a y b son erroneos")
```
