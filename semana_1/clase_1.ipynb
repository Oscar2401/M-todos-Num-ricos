```python
import numpy as np
from pylab import*
from matplotlib import*
import matplotlib.pyplot as plt
```

# Graficar

```python
x=np.linspace(0,20,50) #array de 0 a 20 con 50 datos
y=np.sin(x)
print(x)
print(y)
#python color chart para m√°s colores
plt.plot(x,y,'--*',color = 'salmon',label='$sin(x)$')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend(loc='upper right', shadow=True, fontsize='x-large')
pylab.savefig('sin(x).pdf')
plt.show()
```

#### Ejercicio

```python
x=np.linspace(-5,5,500)
y=np.exp(-x**2)/(np.pi)**(1/2) 
plt.plot(x,y,'-',color = 'magenta',label='$gaussiana$')
plt.xlabel('x',fontsize=18)
plt.ylabel('$\\frac{e^{-x^2}}{\sqrt{\pi}}$',fontsize=18)
plt.legend(loc='upper right', shadow=True, fontsize='x-large')
plt.show()
```

```python
#-pi a pi
x = np.linspace(-np.pi,np.pi,500)
y = np.cos(x) + np.sin(x)
plt.plot(x,y,'-',color = 'navy')
plt.xlabel('x',fontsize=18)
plt.ylabel('cos(x) + sin(x)',fontsize=18)
plt.show()
```
