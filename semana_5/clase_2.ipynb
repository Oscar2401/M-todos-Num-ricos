# Método de Taylor de orden superior y Runge Kutta

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pylab as pl 
from scipy.interpolate import interp1d 
```

Resolveremos una ecuación de la forma:
\begin{equation}
\frac{dy}{dt} = y + t^2 +1
\end{equation}
con condición inicial:
\begin{equation}
y(0) = 0.5 \quad y \quad 0<t<2
\end{equation}

Definimos la función $f(t,y)$ de la ecuación diferencial:

```python
def f(t,y):
    return y-t**2+1
```

Del método de Taylor:
\begin{equation}
\omega_{i+1} = \omega_{i} + hT^{n}(t_i,\omega_i)
\end{equation}


construimos la función $T^{(2)}(t,y)$:

```python
def T_2(t,y,h):
    return f(t,y) + ((h/2)*(f(t,y) -2*t))
```

tambien construimos la función $T^{(4)}(t,y)$:

```python
def T_4(t,y,h):
    return f(t,y)+ ((h/2)*(f(t,y)- 2*t))+ (h**2 / 3)*(f(t,y) - 2*t - 2) + (h**3 / 4)*(f(t,y) - 2*t - 2)
```

Ahora hacemos una función para encontrar la solución con el método de Runge-Kutta de 2do orden:

```python
def RK_2(t,a,b,omg_0,n):
    t = np.linspace(a,b,n)
    omg_rk = [omg_0]
    h = (b-a)/n
    for i in range(1,len(t)):
        omg_rk.append(omg_rk[i-1] + h*f(t[i-1] + h/2, omg_rk[i-1] + (h/2)*(f(t[i-1],omg_rk[i-1]))))
    return [t,omg_rk]
```

Ahora hacemos una función para encontrar la solución con el método de Runge-Kutta de 4do orden:

```python
def RK_4(t,a,b,omg_0,n):
    omg_rk = [omg_0]
    h = (b-a)/n
    t = np.linspace(a,b,n)
    for i in range(1,len(t)):
        k_1 = h*f(t[i-1],omg_rk[i-1])
        k_2 = h*f(t[i-1]+(h/2),omg_rk[i-1]+(k_1/2))
        k_3 = h*f(t[i-1]+(h/2),omg_rk[i-1]+(k_2/2))
        k_4 = h*f(t[i-1]+h,omg_rk[i-1]+k_3)
        omg_rk.append(omg_rk[i-1] + (k_1 + 2*k_2 + 2*k_3 + k_4)/6)
    return [t,omg_rk]
```

```python
n = 5000
a = 0.
b = 2
t = np.linspace(a,b,n)
omg_0 = 0.5
omg_2 = [omg_0]
omg_4 = [omg_0]
h = (b-a)/n
for i in range(1,len(t)):
    omg_2.append(omg_2[i-1] + h*T_2(t[i-1],omg_2[i-1],h))
for i in range(1,len(t)):
    omg_4.append(omg_4[i-1] + h*T_4(t[i-1],omg_4[i-1],h))
```

```python
omg_rk_2 = RK_2(t,a,b,omg_0,n)
omg_rk_4 = RK_4(t,a,b,omg_0,n)
```

La solución exacta es:
\begin{equation}
y(t) = (t+1)^2 - 0.5e^t
\end{equation}

```python
y = (t+1)**2 - (0.5*math.e**t)
```

graficamos:
