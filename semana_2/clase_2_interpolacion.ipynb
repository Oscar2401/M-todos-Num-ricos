```python
import numpy as np
import matplotlib.pyplot as plt
import math
import pylab
```

# Interpolaci√≥n

```python
def interp(x,x_i,y_i):
    L = []
    a = []
    for k in range(len(x_i)):
        numer = np.ones(len(x))
        denom = np.ones(len(x))
        for i in range(len(x_i)):
            if i != k:
                a = x - x_i[i]
                numer = a*numer 
                b = x_i[k] - x_i[i]
                denom = b*denom   
        L.append(np.divide(numer,denom))
    ac = np.zeros(len(L[0]))
    for k in range(len(y_i)):
        ac = ac + L[k]*y_i[k]
    return ac
```

```python
data = np.loadtxt('Datos_prueba-1.txt').T

x_i = data[0][0:7]
y_i = data[1][0:7]

x = np.linspace(np.min(x_i),np.max(x_i),500)
P = interp(x,x_i,y_i)

plt.figure(figsize=(15,10))
plt.plot(x_i,y_i,'ro',label='Datos')
plt.plot(x,P,'k-' , label='Polinomio')

plt.xlabel("x",fontsize=18)
plt.ylabel("f(x)",fontsize=18)
plt.legend(loc='upper right', shadow=True, fontsize='x-large')
plt.ylim(-10,10)
plt.show()
```

```python
data = np.loadtxt('psi_star.dat').T

x_i = data[0]
y_i = data[1]

x = np.linspace(np.min(x_i),np.max(x_i),10000)
P = interp(x,x_i,y_i)

z_1 = 1
z_2 = 4
alf = 3.57731
beta = -2.119E-1
eta = -10
g = -3.9873e3
ro_0 = 1.5649E-2
B = np.power(1+z_1,1-(alf/beta)) 
C = np.power(1+z_2,1-(alf/beta)) 
y = ro_0*np.power(np.power(1+x,alf*eta) + np.power((1+x)/B,beta*eta) + ((1+x)/C)**(g*eta) ,1/eta)

plt.figure(figsize=(15,10))
plt.plot(x_i,y_i,'ro',label='Datos')
plt.plot(x,P,'k-' , label='Polinomio')

plt.plot(x,y,'y-')

plt.xlabel("x",fontsize=18)
plt.ylabel("f(x)",fontsize=18)
plt.legend(loc='upper right', shadow=True, fontsize='x-large')
plt.ylim(-1,1)

plt.show()
```
