puntos: $x_0 = 2$ , $x_1 = 2.5$ y $x_2 = 4$

```python
x = np.linspace(1,5,500)
y = 1/x

P = 0.05*np.power(x,2) - 0.425*x + 1.15

plt.figure(figsize=(10,10))
plt.plot(x,y,'-' , label='$1/x$' )
plt.plot(x,P,'-' , label='Polinomio')

plt.plot(2,1/2,'o')
plt.plot(2.5,1/2.5,'o')
plt.plot(4,1/4,'o')

plt.xlabel("x",fontsize=18)
plt.ylabel("f(x)",fontsize=18)
plt.legend(loc='upper right', shadow=True, fontsize='x-large')
plt.show()
```

```python
def interp(x,x_i,y_i):
    L = []
    a = []
    for k in range(len(x_i)):
        numer = np.ones(len(x))
        denom = np.ones(len(x))
        for i in range(len(x_i)):
            if i != k:
                a = x - x_i[i]
                numer = a*numer 
                b = x_i[k] - x_i[i]
                denom = b*denom   
        L.append(np.divide(numer,denom))
    ac = np.zeros(len(L[0]))
    for k in range(len(y_i)):
        ac = ac + L[k]*y_i[k]
    return ac
```

```python
x = np.linspace(1,6,500)
y = 1/x

x_i = [2,2.5,4,5]
y_i = np.divide(1,x_i)

P = interp(x,x_i,y_i)

plt.figure(figsize=(10,10))
plt.plot(x,y,'-' , label='$1/x$' )
plt.plot(x,P,'-' , label='Polinomio')

plt.plot(x_i,y_i,'o')

plt.xlabel("x",fontsize=18)
plt.ylabel("f(x)",fontsize=18)
plt.legend(loc='upper right', shadow=True, fontsize='x-large')
plt.show()
```

```python
x = np.linspace(0,3,500)
y_i = [0.7651977,0.6200860,0.4554022,0.2818186,0.1103623]
x_i = [1,1.3,1.6, 1.9,2.2]

P = interp(x,x_i,y_i)

plt.figure(figsize=(10,10))
plt.plot(x_i,y_i,'-' , label='$1/x$' )
plt.plot(x,P,'-' , label='Polinomio')

plt.plot(x_i,y_i,'o')

plt.xlabel("x",fontsize=18)
plt.ylabel("f(x)",fontsize=18)
plt.legend(loc='upper right', shadow=True, fontsize='x-large')
plt.show()
```
