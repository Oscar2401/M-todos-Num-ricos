```python
from matplotlib import*
from pylab import*
from scipy.interpolate import interp1d
from scipy.interpolate import CubicSpline
import numpy as np

x=np.array([0.9,1.3,1.9,2.1,2.6,3.0,3.9,4.4,4.7,5.0,6.0,7.0,8.0,9.2,10.5,11.3,11.6,12.0,12.6,13.0,13.3])
fx=np.array([1.3,1.5,1.85,2.1,2.6,2.7,2.4,2.15,2.05,2.1,2.25,2.3,2.25,1.95,1.4,0.9,0.7,0.6,0.5,0.4,0.25])
f_interpolacion=interp1d(x,fx,'cubic')
x_new=linspace(1,12,500)
y=f_interpolacion(x_new)
plot(x_new,y)
show()
#llamamos al arreglo de la interpolacion con un valor(rango)
print(f_interpolacion(2))
print(len(x))

#Cubic spline
cs = CubicSpline(x, fx, bc_type='natural')
xs = np.arange(0.0, 14.0, 0.1)
figure(figsize=(8.5, 3))
plot(x, fx, 'o')
plot(xs, cs(xs))
show()
print(CubicSpline(x, fx, bc_type='natural'))
print(cs(2))
print(len(x))
```
