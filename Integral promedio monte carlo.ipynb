```python
def cn(f,xa,xb,n):
    def func(p):
        return N(f.subs(x,p))
    yp=0.
    xx=[]
    yy=[]
    for i in range(0,n):
        xx.append(rn.uniform(xa,xb))
        yy.append(func(xx[i]))
        yp=yy[i]+yp
    ypp=yp/n
    integral=ypp*(xb-xa)
    #Ireal=N(integrate(func(g),(g,xa,xb)))
    #perror=abs(((integral-Ireal)/Ireal)*100)
    #print("Integral analitica:",Ireal)
    print("Integral obtenida:",integral)
    #print("%error",perror)
    plt.plot(xx,yy,".r")
    
```

```python
cn(x,0.,2.,1000)
```
